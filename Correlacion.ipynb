# Correlación

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
archivo = pd.read_excel("Correlacion.xlsx", sheet_name = "Ingresos")
```

```python
archivo
```

```python
sns.scatterplot(data = archivo, x = "Ingresos", y = "GastosSalidas")
```

```python
np.corrcoef(x= archivo["Ingresos"], y = archivo["GastosSalidas"])
```

```python
archivo.corr()
```

```python
sns.regplot(data = archivo, x = "Ingresos", y = "GastosSalidas", color = "red")
sns.scatterplot(data = archivo, x = "Ingresos", y = "GastosSalidas")
```
